## Read the data file

```python
data_file = pd.read_csv('data-preprocessing/listing-processed.csv', low_memory=False).fillna(0)
data_file = data_file.loc[:, ~data_file.columns.str.contains('^Unnamed')]
data_file.shape
```

```python
data_file
```

## Select features and target for training and testing

#### X is features

```python
X = data_file.drop(
        columns=['id', 'host_id', 'price', 'weekly_price', 'monthly_price']
    )
X.shape
```

#### Y is the target

```python
Y = data_file['price']
Y.shape
```

#### Randomly split data into 80% train and 20% test set

```python
trainX, testX, trainY, testY = \
  model_selection.train_test_split(X, Y, 
  train_size=0.80, test_size=0.20, random_state=1314)

print(trainX.shape)
print(testX.shape)
```

## Principal Component Analysis - PCA

```python
pca = decomposition.PCA(n_components=5)
newTrainX = pca.fit_transform(trainX)
newTestX = pca.transform(testX)
```

```python
newTestX
```

## Decision Tree Classifier

```python
dtclf = tree.DecisionTreeClassifier(random_state=1314, max_depth=10)
```

```python
dtclf.fit(trainX, trainY)
```

```python
dtclf.score(testX, testY)
```

```python
dtclf.feature_importances_
```

## Decision Tree Regressor

```python
dtr = tree.DecisionTreeRegressor(random_state=1314, max_depth=10)
```

```python
dtr.fit(trainX, trainY)
```

```python
dtr.score(testX, testY)
```

## KMeans

```python
Ks = [i + 1 for i in range(5)]
Ks.pop(0)
Ks
```

```python
km = cluster.KMeans(random_state=1314, n_jobs=-1)
```

```python
params = model_selection.ParameterGrid({'n_clusters': Ks})
```

```python
models = []
for p in params:
    km.set_params(**p)
    Y = km.fit_predict(X)
    models.append(Y)
models
```
